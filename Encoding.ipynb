```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
df = sns.load_dataset('tips')
df.head()
```

#### LABEL ENCODING

```python
le = LabelEncoder()
df['encoded_time'] = le.fit_transform(df['time'])
df['time'].value_counts()
```

```python
print("After encoding")
#Dinner=0 ,lunch=1
df['encoded_time'].value_counts()
```

#### ORDINAL ENCODING

```python
oe = OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])
df['encoded_day'] = oe.fit_transform(df[['day']])
df['day'].value_counts()
```

```python
print("After encoding")
df['encoded_day'].value_counts()
```

#### ONE-HOT ENCODING

```python
#no single column encoding in one-hot
# one hot encoding
ohe = OneHotEncoder(sparse_output=False)
encoded_sex = ohe.fit_transform(df[['sex']])

sex_ohe_df = pd.DataFrame(
    encoded_sex,
    columns=ohe.get_feature_names_out(['sex'])
)

df = pd.concat([df.reset_index(drop=True), sex_ohe_df.reset_index(drop=True)], axis=1)
df.head()
```
