```python
#fixing datatypes
df['Date of Purchase'] = pd.to_datetime(df['Date of Purchase'], errors='coerce')
#for both date and time
cols = ['First Response Time', 'Time to Resolution']
df[cols] = df[cols].apply(pd.to_datetime, errors='coerce')
#category
cols = ['Customer Gender','Ticket Type','Ticket Status',
    'Ticket Priority','Ticket Channel','Product Purchased']
df[cols] = df[cols].astype('category')
```

```python
#convert every null value to NaN
df.replace(['NULL', 'None', 'N/A'], np.nan, inplace=True)
```

```python
df.isnull().sum() / len(df) * 100
```

```python
#fill Null values
df['Customer Satisfaction Rating'] = df['Customer Satisfaction Rating'].fillna(0)
df['Time to Resolution'] = df['Time to Resolution'].fillna(0)
df['First Response Time'] = df['First Response Time'].fillna(0)
df['Resolution'] = df['Resolution'].fillna('Not Resolved')
```

```python
df['Customer Gender'] = df['Customer Gender'].str.strip().str.lower()
```

```python
#checking rating does not contain outlier
(~df['Customer Satisfaction Rating'].between(0, 5)).sum()
```

```python
# Fix spelling variations with a mapping
category_map = {
    'tech support': 'Technical Support',
    'Tech Support': 'Technical Support',
    'TECH SUPPORT': 'Technical Support',
    'tech-support': 'Technical Support'
}
df['Product Purchased'] = df['Product Purchased'].replace(category_map)
```
